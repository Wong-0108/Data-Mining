## Data Mining

### Mlxtend

```python
store_data_onehot = pd.get_dummies(store_data, columns=store_data.columns)
display(store_data_onehot)
```

```python
import time

# Start tracking time
start_time = time.time()

#Frequent pattern using aprioro method
fp_apriori = fp.apriori(store_data_onehot,min_support=0.3,use_colnames=True)


# End tracking time
end_time = time.time()

# Display frequent itemsets
display(fp_apriori)

# Calculate execution time
execution_time = end_time - start_time
print(f"Execution time for Apriori: {execution_time:.4f} seconds")
```

```python
import time

# Start tracking time
start_time = time.time()

#frequent pattern using fpgrowth method
fp_fpgrowth = fp.fpgrowth(store_data_onehot, min_support=0.3, use_colnames=True)

# End tracking time
end_time = time.time()

# Display frequent itemsets
display(fp_fpgrowth)

# Calculate execution time
execution_time = end_time - start_time
print(f"Execution time for FP-Growth: {execution_time:.4f} seconds")
```

```python
assoc_rules1 = fp.association_rules(fp_apriori,metric='confidence', min_threshold=0.5,support_only=False)
display(assoc_rules1)
```

```python
assoc_rules2 = fp.association_rules(fp_apriori,metric='lift', min_threshold=0.5,support_only=False)
display(assoc_rules2)
```

```python
assoc_rules3 = fp.association_rules(fp_apriori,metric='conviction', min_threshold=0.5,support_only=False)
display(assoc_rules3)
```
